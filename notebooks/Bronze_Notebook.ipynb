```python
customers_df = spark.read \
    .format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("Files/customers.csv")

display(customers_df)
```

```python
customers_df.write \
    .mode("overwrite") \
    .saveAsTable("dbo.customers")
```
